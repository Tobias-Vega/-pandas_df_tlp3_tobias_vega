```python
import pandas as pd
import sqlite3

# Conexion a la base de datos
conn = sqlite3.connect('Salaries.sqlite')

# Dataframe con consulta SELECT de la tabla salaries
df = pd.read_sql_query('SELECT * FROM salaries', conn)
conn.close()
df.head()
```

1. **Top 10 empleados con mayor salario total (incluyendo beneficios)**
Extrae e imprime un top 10 de expleados con mayor salario incluyendo los beneficios 


```python
# Obteniendo los 10 mayores salarios incluyendo los beneficios
mayores_salarios = df['TotalPayBenefits'].sort_values(ascending=False)
mayores_salarios.head(10)
```

2. **Filtrar empleados con más de 50,000 en horas extra, la columna a tratar es: (OvertimePay)**

```python
# Verificar tipo de dato
df['OvertimePay'].dtype
```

```python
# Convertir a tipo float y en caso de ser strings no númericos reemplazarlos NaN
df['OvertimePay'] = pd.to_numeric(df['OvertimePay'], errors='coerce')

# Pagos de horas extras mayores a 50000
horas_extras = df[df['OvertimePay'] > 50000]
horas_extras
```

3. **Contar cuántos empleados únicos hay por año**

```python
# Se agrupa la cantidad de empleados unicos por año
cantidad_empleados_por_anio = df.groupby('Year')['EmployeeName'].nunique()
cantidad_empleados_por_anio
```

4. **Ver cuántos cargos únicos (JobTitle) existen y los 5 más comunes**

```python
# Contando cantidad de cargos que existen
cantidad_cargos = df['JobTitle'].nunique()
cantidad_cargos
```

```python
# Se muestran los 5 cargos más comunes
cargos_mas_comunes = df['JobTitle'].value_counts().head()
cargos_mas_comunes
```

5. **Mostrar el salario total (TotalPay) promedio por año**

```python
salario_promedio_por_año = df.groupby('Year')["TotalPay"].mean()
salario_promedio_por_año
```

6. **Exporta todos los datos a un CSV y muestra los ultimos 10 valores**

```python
df.to_csv('salaries.csv', index=False)
df.tail(10)
```
